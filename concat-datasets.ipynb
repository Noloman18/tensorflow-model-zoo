```python
import os
import pathlib
import json
import io
import re
import tensorflow as tf
from random import shuffle
from shutil import rmtree

from object_detection.utils import ops as utils_ops
from object_detection.utils import label_map_util

from PIL import Image
from object_detection.utils import dataset_util
from collections import namedtuple, OrderedDict

import xml.etree.ElementTree as ET
from xml.etree.ElementTree import Element
from xml.dom import minidom
```

```python
with open('workspace/road_damage/concat-dataset-config.json', 'r') as f:
    CONFIG = json.load(f)
```

```python
custom_map = {
    'D00':'longitudinal crack',
    'D01':'longitudinal crack',
    'D10':'equal interval lateral crack',
    'D11':'lateral crack',
    'D20':'alligator crack',
    'D40':'pothole',
    'D43':'white blur',
    'D44':'cross walk blur'
}
```

```python
def load_annotation_paths(path:str):
    annotation_paths = list(pathlib.Path(path).glob("*.xml"))
    shuffle(annotation_paths)
    return annotation_paths
```

```python
def load_annotation(fname):
    tree = ET.parse(fname)
    root = tree.getroot()
    return root
```

```python
def get_image_from_annotation(annotation: Element):
    return os.path.join(annotation.find('path').text, annotation.find('filename').text)
```

```python
def update_annotation(annotation: Element, image_base:str):
    path_element = annotation.find('path')
    if path_element is not None:
        path_element.text = image_base
    else:
        path_element = ET.SubElement(annotation, "path")
        path_element.text = image_base
    for member_object in annotation.findall('object'):
        the_name = member_object.find('name')
        try:
            the_name.text = custom_map[the_name.text]
        except:
            annotation.remove(member_object)
    return annotation
```

```python
target_dir = CONFIG['PASCAL_VOC_TARGET_ANNOTATIONS_PATH']
if not os.path.exists(target_dir):
    os.makedirs(target_dir)
    print('CREATED directory',target_dir)
    
data_sets = []
data_sets.append(
    {'annotations_dir':CONFIG['PASCAL_VOC_ANNOTATIONS_PATH_1'], 
     'image_dir':CONFIG['IMAGE_PATH_1']})
data_sets.append(
    {'annotations_dir':CONFIG['PASCAL_VOC_ANNOTATIONS_PATH_2'], 
     'image_dir':CONFIG['IMAGE_PATH_2']})

file_index = 1
for data_set in data_sets:
    annotation_dir = data_set['annotations_dir']
    image_dir = data_set['image_dir']

    annotation_paths = load_annotation_paths(annotation_dir)
    for idx,annotation_path in enumerate(annotation_paths):
        myannotation = load_annotation(annotation_path)
        myannotation = update_annotation(myannotation,image_dir)
        
        new_file = os.path.join(target_dir,'{}.xml'.format(file_index))
        
        with open(new_file,'tw') as f:
            txt = ET.tostring(myannotation, 'utf-8').decode('utf-8').replace('\r\n', '')
            txt = re.sub(r"\s{3,}","",txt)
            f.write(minidom.parseString(txt.encode('utf-8')).toprettyxml(indent="   "))
            print('Finished outputting',new_file)
        file_index += 1
```
